```python
import pandas as pd
```

```python
taxi_zone = pd.read_csv('taxi_zone_lookup.csv')
```

```python
green_trip_data = pd.read_csv('green_tripdata_2019-10.csv')
```

```python
green_trip_data.dtypes
```

```python
green_trip_data.lpep_pickup_datetime = pd.to_datetime(green_trip_data.lpep_pickup_datetime)
green_trip_data.lpep_dropoff_datetime = pd.to_datetime(green_trip_data.lpep_dropoff_datetime)
green_trip_data.store_and_fwd_flag = green_trip_data.store_and_fwd_flag.astype(bool)
green_trip_data.dtypes
```

```python
taxi_zone.dtypes
```

```python
taxi_zone.Borough = taxi_zone.Borough.astype(str)
taxi_zone.Zone = taxi_zone.Zone.astype(str)
taxi_zone.service_zone = taxi_zone.service_zone.astype(str)
```

```python
green_trip_data.dtypes
```

```python
filtered_date = green_trip_data[(green_trip_data.lpep_pickup_datetime >= '2019-10-01 00:00:00') & (green_trip_data.lpep_dropoff_datetime < '2019-11-01 00:00:00')
]
```

```python
up_1_mile = filtered_date[filtered_date.trip_distance <= 1]
up_1_mile
```

```python
one_to_three_mile = filtered_date[(filtered_date.trip_distance > 1) & (filtered_date.trip_distance <= 3) ]
one_to_three_mile
```

```python
green_trip_data['duration'] = green_trip_data.lpep_dropoff_datetime - green_trip_data.lpep_pickup_datetime
trip_with_duration = green_trip_data
```

```python
trip_with_duration['pickup_date'] = trip_with_duration['lpep_pickup_datetime'].dt.date
trip_with_duration.groupby('pickup_date')['trip_distance'].max().sort_values()
```

```python
taxi_zone_sub = taxi_zone[['LocationID','Zone']]
green_trip_data.pickup_date = green_trip_data.pickup_date.astype(str)
trip_191018 = green_trip_data[green_trip_data['pickup_date'] == '2019-10-18']
top_pickup_locations = trip_191018.merge(taxi_zone_sub,left_on='PULocationID',right_on='LocationID')[['Zone','total_amount']]
```

```python
top_pickup_locations.groupby('Zone')['total_amount'].sum().sort_values()
```

```python
largest_tip = green_trip_data.merge(taxi_zone_sub,left_on='PULocationID',right_on='LocationID')
```

```python
largest_tip = largest_tip[largest_tip['Zone'] == "East Harlem North"][["DOLocationID","tip_amount"]]
```

```python
largest_tip = largest_tip.groupby('tip_amount').max()
largest_tip.merge(taxi_zone_sub,left_on='DOLocationID',right_on='LocationID')
```
